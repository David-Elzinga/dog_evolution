```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
```

```python
with open("data/efast/master_df_parsed_15.pickle", "rb") as f:
    df = pickle.load(f)

df['speciation_bool'] = df['speciation_ybp'].notna()
```

```python
df['speciation_bool'].mean()
```

```python
df.groupby(['food_scheme'])['speciation_bool'].mean()
```

```python
const_df = df[df['food_scheme'] == 'constant']
incr_df = df[df['food_scheme'] == 'increasing']

fig, ax = plt.subplots(1, 1)

ax.hist(const_df.groupby(['param_set'])['speciation_bool'].mean(), 
        bins = [0, 0.05, 0.1, 0.2, 0.4, 0.6, 0.8, 1],
        color = 'red',
        alpha = 0.25,
        label = 'Constant Human Food'
       )
ax.hist(incr_df.groupby(['param_set'])['speciation_bool'].mean(),
        bins = [0, 0.05, 0.1, 0.2, 0.4, 0.6, 0.8, 1], 
        color = 'blue', 
        alpha = 0.25,
        label = 'Increasing Human Food'
       )

ax.legend(fontsize = 14)
ax.set_xlabel('Speciation Proportion', fontsize = 18, labelpad = 10)
ax.set_ylabel('Parameter Sets', fontsize = 18, labelpad = 10)
fig.savefig('figures/distributions/speciation_proportion_15.pdf', bbox_inches='tight')
```
